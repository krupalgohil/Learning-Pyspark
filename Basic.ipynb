```python
from pyspark.sql import SparkSession 

spark = SparkSession.builder.appName("BigMart").getOrCreate()

print("spark started")
```

## Load data 


```python
df = spark.read.format('csv')\
    .option('inferSchema',True)\
    .option('header',True)\
    .load('BigMart Sales.csv')
```

```python
df.show(20,truncate=False)
```

### SCHEMA - DDL and StructType()

```python
df.printSchema()
```

#### DDL SCHEMA 

```python
my_ddl_shcema = '''  
            Item_Identifier STRING,
            Item_Weight STRING,
            Item_Fat_Content STRING,
            Item_Visibility DOUBLE,
            Item_Type STRING,
            Item_MRP DOUBLE,
            Outlet_Identifier STRING,
            Outlet_Establishment_Year INT,
            Outlet_Size STRING,
            Outlet_Location_Type STRING, 
            Outlet_Type STRING,
            Item_Outlet_Sales DOUBLE 
'''
```

```python
df1 = spark.read.format('csv')\
    .schema(my_ddl_shcema)\
    .option('header',True)\
    .load('BigMart Sales.csv')
```

```python
df1.show()
```

```python
df1.printSchema()
```

#### StructType Schema 

```python
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```python
my_strct_schema = StructType([ 
    StructField('Item_Identifier',StringType(),True), 
    StructField('Item_Weight',StringType(),True),
    StructField('Item_Fat_Content',StringType(),True), 
    StructField('Item_Visibility',StringType(),True), 
    StructField('Item_MRP',StringType(),True), 
    StructField('Outlet_Identifier',StringType(),True), 
    StructField('Outlet_Establishment_Year',StringType(),True), 
    StructField('Outlet_Size',StringType(),True), 
    StructField('Outlet_Location_Type',StringType(),True), 
    StructField('Outlet_Type',StringType(),True), 
    StructField('Item_Outlet_Sales',StringType(),True)
])
```

```python
df2 = spark.read.format('csv')\
    .schema(my_strct_schema)\
    .option('header',True)\
    .load('BigMart Sales.csv')

df2.printSchema()
```

```python
df.filter(col('Item_Fat_Content') == 'Regular').show(5)
```

#### Scenrio - 2

```python
df.filter((col('Item_Fat_Content') == 'Regular') & (col('Item_Weight')<10)).show(5)
```
