```python
import os
import sys

os.environ['PYSPARK_PYTHON'] = sys.executable
os.environ['PYSPARK_DRIVER_PYTHON'] = sys.executable
```

```python
from pyspark.sql import SparkSession 

spark = SparkSession.builder.appName("BigMart").getOrCreate()

print("spark started") 
```

## Load data 


```python
df = spark.read.format('csv')\
    .option('inferSchema',True)\
    .option('header',True)\
    .load('BigMart Sales.csv')
```

```python
df.show(20,truncate=False)
```

### SCHEMA - DDL and StructType()

```python
df.printSchema()
```

#### DDL SCHEMA 

```python
my_ddl_shcema = '''  
            Item_Identifier STRING,
            Item_Weight STRING,
            Item_Fat_Content STRING,
            Item_Visibility DOUBLE,
            Item_Type STRING,
            Item_MRP DOUBLE,
            Outlet_Identifier STRING,
            Outlet_Establishment_Year INT,
            Outlet_Size STRING,
            Outlet_Location_Type STRING, 
            Outlet_Type STRING,
            Item_Outlet_Sales DOUBLE 
'''
```

```python
df1 = spark.read.format('csv')\
    .schema(my_ddl_shcema)\
    .option('header',True)\
    .load('BigMart Sales.csv')
```

```python
df1.show()
```

```python
df1.printSchema()
```

#### StructType Schema 

```python
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```python
my_strct_schema = StructType([ 
    StructField('Item_Identifier',StringType(),True), 
    StructField('Item_Weight',StringType(),True),
    StructField('Item_Fat_Content',StringType(),True), 
    StructField('Item_Visibility',StringType(),True), 
    StructField('Item_MRP',StringType(),True), 
    StructField('Outlet_Identifier',StringType(),True), 
    StructField('Outlet_Establishment_Year',StringType(),True), 
    StructField('Outlet_Size',StringType(),True), 
    StructField('Outlet_Location_Type',StringType(),True), 
    StructField('Outlet_Type',StringType(),True), 
    StructField('Item_Outlet_Sales',StringType(),True)
])
```

```python
df2 = spark.read.format('csv')\
    .schema(my_strct_schema)\
    .option('header',True)\
    .load('BigMart Sales.csv')

df2.printSchema()
```

## TRANSFORMATIONS

#### SELECT

```python
df.show()
```

```python
df.select(col('Item_Identifier'),col('Item_Identifier')).show(5)
```

### ALIAS

```python
df.select(col('Item_Identifier').alias('item_id')).show(5)
```

```python
df.show(5)
```

### FILTER

#### Scenario - 1

```python
df.filter(col('Item_Fat_Content') == 'Regular').show(5)
```

#### Scenrio - 2

```python
df.filter((col('Item_Fat_Content') == 'Regular') & (col('Item_Weight')<10)).show(5)
```

#### Scenario - 3

```python
df.filter((col('Outlet_Size').isNull()) & (col('Outlet_Location_Type').isin('Tier 1','Tier 2'))).show(5)
 
```

### withColumnRenamed

```python
df.withColumnRenamed('Item_Weight','Item_Wt').show(1)
```

### withColumn 

#### Scenario - 1 

```python
df = df.withColumn('flag',lit('new'))
df.show(5)
```

```python
df.withColumn('multiply',col('Item_weight')*col('Item_MRP')).show(5)
```

#### Scenario - 2


```python
df = df.withColumn('Item_Fat_Content',regexp_replace(col('Item_Fat_Content'),'Regular','Reg'))\
    .withColumn('Item_Fat_Content',regexp_replace(col('Item_Fat_Content'),'Low Fat','LF'))
```

```python
df.show(5)
```

### Type Casting

```python
df = df.withColumn('Item_Weight', col('Item_Weight').cast(StringType()))
df.printSchema()
```

### sort

#### Scenario - 1

```python
df.sort(col('Item_Weight').desc()).show(5)
```

#### Scenario - 2


```python
df.sort(col('Item_Visibility').asc()).show(5)
```

#### Scenario - 3

```python
df.sort(['Item_weight','Item_Visibility'],ascending=[0,0]).show(5)
```

#### Scenario - 4

```python
df.sort(['Item_weight','Item_Visibility'],ascending=[0,1]).show(5)
```
